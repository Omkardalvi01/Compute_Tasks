```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
plt.style.use('ggplot')
pd.set_option('display.max_columns', 200)
pd.options.display.max_info_columns = 1000
```

```python
data = pd.read_csv('data.csv')
data
```

```python
data.dropna(axis=0)
#Every player has atleast one missing values, makes more sense to remove columns
```

```python
#Removing missing values 
data.dropna(inplace=True, axis=1)
data
```

```python
#Data with no missing values
data.info()
```

```python
#chaning to correct format
data['dob'] = pd.to_datetime(data['dob'])
data.info()
```

```python
#Checking for duplicates
data.duplicated().sum()
```

```python
#Double checking, data does not seem to have any duplicate values
data.loc[data.duplicated(subset=['long_name','age','nationality'])]
```

```python
data.describe()
```

```python
ax = data['age'].plot(kind='hist',figsize=(15,5), bins= 30)
ax.set_title('Age distribution of football players')
ax.set_xlabel('Age')
plt.show()
```

```python
ax = data['wage_eur'].value_counts().head(20).plot(kind='bar',figsize=(15,5))
ax.set_title('Most Common Wage for Players')
ax.set_xlabel('Wage(Euros)')
ax.set_ylabel('Frequency')
plt.show()
```

```python

top12 = data.query('wage_eur > 260000')
top12.value_counts().sum()
#top 12 player make more than 260k euros
```

```python
plt.figure(figsize=(15,8))
ax = sns.barplot(data=top12, x='short_name', y='wage_eur')
ax.set_title("Wage Distribution among the most paid players")
plt.show()
#here we can see who are the top 12 most players in fifa 
```

```python
plt.figure(figsize=(15,8))
ax = sns.barplot(data=top12, x='short_name', y='value_eur')
ax.set_title("Value Distribution among the most paid players")
plt.show()
#shows that wage of player is not indicative of their value
```

```python
preferred_foot_counts = data['preferred_foot'].value_counts()
plt.pie(preferred_foot_counts, labels=preferred_foot_counts.index)
plt.title("Distribution of Preferred Foot (Right vs Left)")
plt.show()
```

```python
Nationality_counts = data['nationality'].value_counts().head(10)
plt.figure(figsize=(8,8))
plt.pie(Nationality_counts, labels=Nationality_counts.index)
plt.title("Top 10 Most Represented Nationalities in the Player Dataset")
plt.show()
#Distribution of countries with highest number of professional players, england is the highest
```

```python
sns.pairplot(data=data, vars=['age','potential','wage_eur', 'value_eur','international_reputation'],hue='preferred_foot')
plt.show()
```

```python
corr = data[['potential', 'value_eur', 'wage_eur', 'overall', 'age']].corr()
sns.heatmap(corr)
plt.show()
```
